```r
# plot the number of vaccines given vs time for different manufacturer
options(repr.plot.width=25, repr.plot.height=8)  
italy_man_plot <- ggplot(data=italy_vaccinations_df, aes(x=date, y=total_vaccinations, color=vaccine)) +   
     theme_linedraw() +                                            
     geom_line(lwd=1.5) +                                                             
     theme(text = element_text(size=20)) +
     ylab("Number of vaccines given") +
     ggtitle("Number of vaccines given vs time - Italy") +
     theme(plot.title = element_text(hjust = 0.5),
           axis.text.x = element_text(angle = 60, hjust = 1)) +
     scale_x_date(date_breaks = "1 months", date_labels="%b-%y") +
     scale_y_continuous(trans='log10')                               

# plot total number of vaccines shot per day 

#total_vaccines_per_day = aggregate(italy_vaccinations_df$total_vaccinations, list(italy_vaccinations_df$date), sum)[[2]]
#vaccines_per_day       = diff(total_vaccines_per_day)

# the follownig routine fails due to no data for astrazeneca after Jan 22, done with for loops although is highly inefficient

vaccines_per_day <- rep(0, length(unique(italy_vaccinations_df$date)))
for (i in 1:length(unique(italy_vaccinations_df$date))){
    date_ = unique(italy_vaccinations_df$date)[[i]]
    for (vaccine_ in unique(italy_vaccinations_df$vaccine)){
        current_shots  = filter(italy_vaccinations_df, date==date_ & vaccine==vaccine_)
        previous_shots = filter(italy_vaccinations_df, date==(date_-1) & vaccine==vaccine_)
        if (dim(current_shots)[1]!=0){
            if (dim(previous_shots)[1]!=0){
                vaccines_per_day[i] <- vaccines_per_day[i] + current_shots$total_vaccinations - previous_shots$total_vaccinations
            }
        }
    }
}

italy_day_plot <- ggplot() +   
     theme_linedraw() +                                            
     geom_line(aes(x=unique(italy_vaccinations_df$date), y=vaccines_per_day), 
               lwd=1.5,
               color="cornflowerblue") +                                                             
     theme(text = element_text(size=20)) +
     ylab("Total number of vaccines given per day") +
     xlab("date") +
     ggtitle("Total number of vaccines given per day vs time - Italy") +
     theme(plot.title = element_text(hjust = 0.5),
           axis.text.x = element_text(angle = 60, hjust = 1)) +
     scale_x_date(date_breaks = "1 months", date_labels="%b-%y")                                  

grid.arrange (italy_man_plot , italy_day_plot ,  nrow=1, widths=c(12, 10))
```

Notice that around Mar -22 there is an unphysical record for USA - Johnson&Johnson that corrupts the daily vaccinations visualizations -> we have a negative difference in total number of vaccines given... that's impossible! We can try fix that point.

```r
# detect broken record date given the previous information of the plot
usa_vaccinations_df = filter(vaccinations_df, location=="United States")
df = usa_vaccinations_df[usa_vaccinations_df$date > as.Date("2022-03-01"), ]
for (i in 1:length(df$date)){
    if(df[[i, "total_vaccinations"]] < mean(df[df$vaccine == "Johnson&Johnson", ]$total_vaccinations)){
        date_ = df[[i, "date"]]
        }
}

# get broken record index and replace the broken record with it previous value
broken_record = which(usa_vaccinations_df$date == date_ & usa_vaccinations_df$vaccine == "Johnson&Johnson")
usa_vaccinations_df[broken_record, ] = usa_vaccinations_df[usa_vaccinations_df$date == date_-1 & usa_vaccinations_df$vaccine == "Johnson&Johnson", ]
```

```r
# filter to detect unespected behaviour in daily vaccination plot and artificailly repair those records
daily_vacc_filter <-function(difference){
    for (i in 1:length(difference)){
        if (difference[i]<0){
            difference[i]=0
        }
        if (difference[i]>10*mean(difference)){
            difference[i]=0
        }
    }
    return(difference)
    }
```

```r
usa_man_plot <- ggplot(data=usa_vaccinations_df, aes(x=date, y=total_vaccinations, color=vaccine)) +   
     theme_linedraw() +                                            
     geom_line(lwd=1.5) +                                                             
     theme(text = element_text(size=20)) +
     ylab("Number of vaccines given") +
     ggtitle("Number of vaccines given vs time - USA") +
     theme(plot.title = element_text(hjust = 0.5),
           axis.text.x = element_text(angle = 60, hjust = 1)) +
     scale_x_date(date_breaks = "1 months", date_labels="%b-%y") +
     scale_y_continuous(trans='log10')  
     
total_vaccines_per_day = aggregate(usa_vaccinations_df$total_vaccinations, 
                                   list(usa_vaccinations_df$date), 
                                   sum)[[2]]
vaccines_per_day       = daily_vacc_filter(diff(total_vaccines_per_day))
usa_day_plot <- ggplot() +   
     theme_linedraw() +                                            
     geom_line(aes(x=unique(usa_vaccinations_df$date)[-1], y=vaccines_per_day), 
               lwd=1.5,
               color="cornflowerblue") +                                                             
     theme(text = element_text(size=20)) +
     ylab("Total number of vaccines given per day") +
     xlab("date") +
     ggtitle("Total number of vaccines given per day vs time - United States") +
     theme(plot.title = element_text(hjust = 0.5),
           axis.text.x = element_text(angle = 60, hjust = 1)) +
     scale_x_date(date_breaks = "1 months", date_labels="%b-%y")                                  

grid.arrange(usa_man_plot , usa_day_plot ,  nrow=1, widths=c(12, 10))
```
